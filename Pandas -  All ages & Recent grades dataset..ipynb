```python
import pandas as pd
```

```python
all_ages = pd.read_csv("all_ages.csv")
recent_grades = pd.read_csv("recent_grades.csv")
```

```python
all_ages.head()
```

```python
recent_grades.head()
```

```python
aa_major_cat = all_ages["Major_category"].unique()
rg_major_cat = recent_grades["Major_category"].unique()

aa_cat_count = {}
rg_cat_count = {}

def count_student_major_cat(major_cat, df, cat_count):
    for category in major_cat:
        subset = df[df["Major_category"] == category]
        nbr_of_students = subset["Total"].sum()
        cat_count[category] = nbr_of_students
        
```

```python
count_student_major_cat(aa_major_cat, all_ages, aa_cat_count)
aa_cat_count
```

```python
import numpy
aa_students_per_major_cat = all_ages.pivot_table(index="Major_category", values="Total", aggfunc=numpy.sum)
aa_students_per_major_cat
```

```python
count_student_major_cat(rg_major_cat, recent_grades, rg_cat_count)
rg_cat_count
```

```python
rg_students_per_major_cat = recent_grades.pivot_table(index="Major_category", values="Total", aggfunc=numpy.sum)
rg_students_per_major_cat
```

```python
low_wage_proportion = (recent_grades["Low_wage_jobs"].sum() / recent_grades["Total"].sum()) * 100
print(f"The proportion of low wage jobs is : {low_wage_proportion}")
```

How much recent grades performs worste than the all population.

```python
rg_lower_count = 0

for major in recent_grades["Major"][recent_grades["Major"].isnull() == False].unique():
 
    rg_unemployed_rate = recent_grades[recent_grades["Major"] == major]["Unemployment_rate"].iloc[0]
    aa_unemployed_rate = all_ages[all_ages["Major"] == major]["Unemployment_rate"].iloc[0]

    if rg_unemployed_rate < aa_unemployed_rate:
        rg_lower_count += 1

rg_lower_count
```
